Orange es una caja de herramientas de minería y visualización de datos interactiva para principiantes y expertos por igual. Para explorar datos con Orange, no se requiere programación ni conocimientos matemáticos profundos. Se basa en la creencia que las herramientas de ciencia de datos basadas en flujos de trabajo democratizan la ciencia de datos ocultando mecanismos subyacentes complejos y exponiendo conceptos intuitivos. Cualquiera que posea datos, o esté motivado para echar un vistazo a los datos, debería tener los medios para hacerlo.

Sin título.jpg

Orange es un programa informático para realizar minería de datos y análisis predictivo desarrollado en la facultad de informática de la Universidad de Ljubljana(Eslovenia). Consta de una serie de componentes desarrollados en C++ que implementan algoritmos de minería de datos, así como operaciones de preprocesamiento y representación gráfica de datos.

Los componentes de Orange pueden ser manipulados desde programas desarrollados en Python o a través de un entorno gráfico.

Se distribuye bajo licencia GPL.

Orange es un software de minería de datos basado en componentes. Incluye una variedad de técnicas de visualización, exploración, preprocesamiento y modelado de datos. Se puede utilizar a través de una interfaz de usuario agradable e intuitiva o, para usuarios más avanzados, como un módulo para el lenguaje de programación Python.
